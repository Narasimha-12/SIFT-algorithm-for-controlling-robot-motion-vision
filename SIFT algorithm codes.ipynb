1.Keypoints identification

```python
import cv2 
import matplotlib.pyplot as plt

# Reading image
img1 = cv2.imread(r"testing gesture.webp")  
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# Keypoints
orb = cv2.ORB_create()
keypoints_1, descriptors_1 = orb.detectAndCompute(gray1, None)

img_1 = cv2.drawKeypoints(gray1, keypoints_1, img1)
print("The image is converted into grey scale(black and white) and also it is transformed into rotation invariant.")
print("The coloured dots in the image represent the KEY POINTS of the image.")
plt.imshow(img_1)
plt.show()
print("ZOOM IN THE IMAGE TO SEE THE KEY POINTS")
```

2.Transformation of testing gesture and trained gestures by a robot

```python
import cv2 
import matplotlib.pyplot as plt

# read images
img1 = cv2.imread(r"testing gesture.webp")  
img2 = cv2.imread(r"trained images for a robot.png") 

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

figure, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].imshow(img1, cmap='gray')
ax[1].imshow(img2, cmap='gray')
plt.show()
```

3.Gesture Matching

```python
import cv2 
import matplotlib.pyplot as plt

# read images
img1 = cv2.imread(r"testing gesture.webp")  
img2 = cv2.imread(r"trained images for a robot.png") 

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# sift
sift = cv2.SIFT_create()

keypoints_1, descriptors_1 = sift.detectAndCompute(img1, None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2, None)

# feature matching
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1, descriptors_2)
matches = sorted(matches, key=lambda x: x.distance)

img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)
plt.imshow(img3)
plt.show()
print("We can see the key points matching between the gesture shown by user and the gestures trained in the robot.")
print("Action related to gesture of highest key points matching will be performed by the robot.")
```
